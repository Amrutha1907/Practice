```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset('iris')
iris
```

```python
iris.columns
```

```python
SL=iris['sepal_length'].mean()
print('mean of sepal_length=',SL)
```

```python
SW=iris['sepal_width'].mean()
print('mean of sepal_width=',SW)
```

```python
PL=iris['petal_length'].mean()
print('mean of petal_length =',PL)
```

```python
PW=iris['petal_width'].mean()
print('mean of petal_width=',PW)
```

```python
iris.isna().sum()
```

```python
#scatter plot
import matplotlib.pyplot as plt
plt.scatter(iris['species'],iris['sepal_length'],s=10,c='green',marker='*')
plt.title(' species vs SL',fontsize=15)
```

```python
iris[['sepal_width','petal_length']].plot.hist()
```

```python
plt.figure(figsize=(8,8))
plt.boxplot([iris['sepal_length'],iris['sepal_width'],iris['petal_length'],iris['petal_width']])
plt.title('Box plot representation of dataset iris',fontsize=14)
plt.xticks([1,2,3,4],['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```

```python
#pair plots
pairplot_iris=iris[['sepal_length','sepal_width','petal_length','petal_width']]
sns.pairplot(pairplot_iris)
```
