# 3. Whether the same amount was spent for the three advertisements TV , Radio and Newspaper

```python
#Let the hypothesis are
```

H0= Same amount spent for the three advertisements,µ1=µ2=µ3
Ha= Amount spent for the advertisements are not same , μ1,µ2 and µ3 are not equal


```python
# significance level
```

significance level, α=0.05 

```python
import numpy as np
from scipy.stats import f_oneway
x=np.array(data['TV'])
y=np.array(data['Radio'])
z=np.array(data['Newspaper'])
f_oneway(x,y,z)
```

```python
# comparing the p value with significance level 0.05
```

p value: 4.552931539744962e-103 ≈ 4.55 x 10^-103 < 0.05

so that Rejecting the null hypothesis
ie,the Amount spent for the advertisements are not same , μ1,µ2 and µ3 are not equal.
