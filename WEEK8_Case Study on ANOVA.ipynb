# Check whether all the zones generate the same amount of sales.

```python
# we have to check one factor sales over 4 different Zones so we have to use ANOVA.
```

```python
#Lets the hypothesis is 
```

#null hypothesis
H0:All Zones generate same amount of sales
#alternative hypothesis
Ha:All zones generate different amount of sales


#significance level
significance level, α=0.05

```python
import numpy as np
from scipy.stats import f_oneway
a=np.array(data['Zone - A'])
b=np.array(data['Zone - B'])
c=np.array(data['Zone - C'])
d=np.array(data['Zone - D'])
f_oneway(a,b,c,d)
```

```python
#conclusion
```

the p value is lessthan that of significance level α,
pvalue<α
rejecting the null hypothesis ie,All zones generate different amount of sales
